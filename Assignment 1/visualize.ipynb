```python
import numpy as np
import matplotlib.pyplot as plt

def plot_2d_array(arr, title="2D Array", cmap="viridis"):
    plt.figure()
    plt.imshow(arr, origin='lower', aspect='auto', cmap=cmap)
    plt.colorbar(label='Value')
    plt.title(title)
    plt.xlabel("x index")
    plt.ylabel("y index")
    plt.show()
```

```python
u = np.zeros((50, 50))
u[0, :] = 1  # Top boundary
u[-1, :] = 0  # Bottom boundary
plot_2d_array(u)
```
